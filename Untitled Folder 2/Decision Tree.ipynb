### Decision Tree:
  
  
  Decision Tree is a tree structure algorithm.it is also known as CART(Classification and Regression Tree)
  
  
  it use the decision (rules) for classifing data
  
  it use the entropy value for taking decision

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
from sklearn.datasets import load_iris
d=load_iris()
df=pd.DataFrame(d.data,columns=d.feature_names)
df['Target']=d.target
df.head()
```

```python
#x=df.drop('Target,axis'=1)
x=df.iloc[:,:-1]
y=df['Target']
```

```python
# import decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pre=model.predict(x_test)
accuracy_score(y_test,y_pre)
```

```python
from sklearn.datasets import load_diabetes
d=load_diabetes()
df=pd.DataFrame(d.data,columns=d.feature_names)
df['Target']=d.target
df
```

```python
x=df.drop('Target',axis=1)
y=df['Target']
```

```python
from sklearn.tree import DecisionTreeRegressor
d1=DecisionTreeRegressor()
d1.fit(x,y)
d1.score(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
d1.fit(x_train,y_train)
d1.score(x_train,y_train)
```

```python
# find the predict score
y_pred=d1.predict(x_test)
s=pd.DataFrame({'a':y_test,'p':y_pred})
s.head()
```

```python
# find the error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
mean_squared_error(y_test,y_pred)
mean_absolute_error(y_test,y_pred)
np.sqrt(mean_squared_error(y_test,y_pred))
```
